```python
# finding Phone numbers pattern in string
```

```python
def match_phone_number(phone_number):
    pattern = re.compile(r'''
        (                   # Start of capturing group for the entire phone number
            \+\d{1,2}       # Match the international dialing code (e.g., +1, +44)
            |               # OR
            \d{1,4}         # Match the optional area code (1 to 4 digits)
        )?                  # End of the optional capturing group
        [\s.-]?             # Match a whitespace character, dot, or hyphen (optional)
        (\d{1,})            # Match one or more digits for the main part of the number
        [\s.-]?             # Match a whitespace character, dot, or hyphen (optional)
        (\d{1,})            # Match one or more digits for the second part of the number
        [\s.-]?             # Match a whitespace character, dot, or hyphen (optional)
        (\d{1,})            # Match one or more digits for the third part of the number
        ''', re.VERBOSE)

    match = re.match(pattern, phone_number)
    return bool(match)

phone_number1 = "+1 555-123-4567"
phone_number2 = "123.456.7890"
phone_number3 = "9876543210"

print(match_phone_number(phone_number1))  # True
print(match_phone_number(phone_number2))  # True
print(match_phone_number(phone_number3))  # True
```

```python
def match_email_address(email):
    pattern = re.compile(r'''
        ^                           # Start of the string
        [a-zA-Z0-9._%+-]+           # Username: one or more alphanumeric characters, dots, underscores, percent signs, plus signs, or hyphens
        @                           # @ symbol
        [a-zA-Z0-9.-]+              # Domain name: one or more alphanumeric characters, dots, or hyphens
        \.[a-zA-Z]{2,}              # Top-level domain: at least 2 characters
        $                           # End of the string
        ''', re.VERBOSE)

    match = re.match(pattern, email)
    return bool(match)

email1 = "user@example.com"
email2 = "user123@my-domain.net"
email3 = "invalid.email"

print(match_email_address(email1))  # True
print(match_email_address(email2))  # True
print(match_email_address(email3))  # False
```

```python
# Common Characters
```

```python
'''Literal Characters:

Characters that match themselves.
Example: a matches the character 'a'.
Metacharacters:

Characters with special meanings in regular expressions.
Common metacharacters include:
.: Matches any character except a newline.
^: Anchors the regex at the start of the string.
$: Anchors the regex at the end of the string.
Character Classes:

Enclosed in square brackets ([]).
Example: [aeiou] matches any vowel.
Special characters within brackets:
^ as the first character negates the class. [^aeiou] matches any character except vowels.
Ranges:

Inside character classes, you can specify a range of characters.
Example: [a-z] matches any lowercase letter.
Quantifiers:

Specify the number of occurrences.
Common quantifiers include:
*: Matches 0 or more occurrences.
+: Matches 1 or more occurrences.
?: Matches 0 or 1 occurrence.
{n}: Matches exactly n occurrences.
{n,}: Matches n or more occurrences.
{n,m}: Matches between n and m occurrences.
Escape Character:

\ is used to escape a metacharacter, allowing you to match it as a literal character.
Example: \. matches a literal dot.
Alternation:

| is used for alternation, allowing you to match one of several patterns.
Example: cat|dog matches either 'cat' or 'dog'.
Grouping:

() is used to group patterns together.
Example: (ab)+ matches one or more occurrences of 'ab'.
Wildcard:

.* matches any sequence of characters (except newline).
.*? is a non-greedy version of .*.
Anchors:

\b matches a word boundary.
\B matches a non-word boundary.'''
```
